***Modelo de regresión lineal simple***

Para analizar un poco más a detalle el uso de un indicador económico como lo es el **GDP**, se utilizará como variable explicativa del nivel de felicidad. Donde, a pesar de haber identificado previamente una correlación medianamente positiva, es pertinente profundizar al análisis para evaluar más a detalle la fuerza de la asociación. Utilizando el logaritmo del GDP para observar relaciones más lineales.

De esta forma, se planteo un modelo de regresión lineal simple que permitirá estimar el efecto del GDP sobre la felicidad y su significancia estadística de la asociación, a través de la siguiente ecuación:

Felicidad_i= B_0 + B_1 * logaritmo GDP






Utilizando la función antes descrita `np.log()` para la trasformación del GDP y `LinealRegression` para ajustar un modelo lineal simple que relaciona  `Logaritmo GDP` con `Felicidad`. Donde, con el uso de `fit()` se estiman los coeficiantes para así visualizarlo en un gráfico mediante `plt.plot` y `plt.scatter`.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data["LogGDP"] = np.log(data["GDP"])
X = data[["LogGDP"]]
y = data["Felicidad"]

#Modelo lineal
modelo = LinearRegression().fit(X, y)

# Gráfico
plt.scatter(X, y, alpha=0.6, label="Países")
plt.plot(X, modelo.predict(X), color="red", label="Regresión lineal")
plt.xlabel("Logaritmo GDP")
plt.ylabel("Felicidad")
plt.title("Felicidad vs Logaritmo GDP")
plt.legend()
plt.show()
```

De esta forma, al usar la función `OLS` perteneciente a statsmodels es posibe ajustar el modelo de regresión lineal simple entre las variables. Imprimiendo el `p-value` y ` R²`, para así analizar si la asociación es significativa que tanto explica la felicidad.

```python

import statsmodels.api as sm

print("Intercección B_0:", modelo.intercept_)
print("Pendiente B_1:", modelo.coef_[0])

modelo = sm.OLS(y, X).fit()
print("p-value:", modelo.pvalues["LogGDP"])
print("R²:", modelo.rsquared)
```

Mostrando a través del modelo y su pendiente de 0.27, un efecto positivo entre el GDP y Felicidad. Donde, con un p-value pequeño se determina que la relación es estadísticamente significativa. Además de señalar con el R² de 0.97 que el GDP explica gran parte de la variabilidad de la felicidad.

Comprendiendo que el GDP mantiene una relación importante con la felicidad en la mayoría de los casos; no obstante, reconociendo que existen otros factores que Influyen directamente en el bienestar de la población.
