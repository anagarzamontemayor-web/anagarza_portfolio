***Reflexión sobre el diseño del estudio***

Una variable adicional la cual sería relevante medir serían las *horas de actividad física*. El ejercicio, además de contribuir a mantenerte en tu peso ideal, reduce el riesgo de contraer enfermedades y aporta beneficios a corto y  largo plazo. [1]

Análizar esta variable permitiría contar con una perspectiva más detallada de los individuos y comprender como se relaciona esta variable con su diagnóstico general. Tratandose de una variable de naturaleza cuantitativa, al expresarse en número de horas semanales o diarias. Facilitando el estudio de patrones y tendencias que ofrecerín una visión más completa de los factores que influyen en la salúd nutricional.

Por su parte, una variable adicional, obtenida con los datos previamente brindados, que sería relevante para el presente estudio, es el *índice de masa corporal (IMC)*. Este, permite evaluar y medir el peso corporal de una persona en relación con su estatura. [2]

Consolidándose como un indicador clave en la salud pública y medicina preventiva, el IMC es considerado como un estándar para evaluar el riesgo vinculado al bajo peso, sobrepeso y obesidad en adultos, según la Organización Mundial de la Salud (OMS). [2]

De tal manera que, los resultados obtenidos se encuentran clasificados mediante distintos rangos previamente estandarizados según la OMS, como se observa a continuación: [2]

*  Bajo peso: Menos que 18.5

*  Normalidad: 18.5 a 24.9
*  Sobrepeso: 25 - 29.9
*  Obesidad: 30 o más

La estimación categórica actual del peso implica múltiples consideraciones. No obstante, es importante destacar que dicha medida no constituye un diagnóstico de salud preciso, ya que no distingue entre masa muscular y grasa corporal. Antes bien, funciona como indicador preventivo del riesgo de desarrollo de enfermedades crónicas y sus derivados, tales como diabetes tipo 2, enfermedades del corazón y necesidad de intervenciones específicas. [2]

Dicha medida se define de naturaleza cuantitativa al calcularse matemáticamente un número real perteneciente a una categoría. Para ello, se divide el peso del individuo (kg) entre el cuadrado de su altura (m²), de la siguiente forma: [2]




```python
data['IMC'] = data['Peso'] / (data['Estatura']**2)
```

Así, una vez teniendo los valores que representan el `IMC`, es posible clasificar los valores numéricos con la función `pd.cut()`, limitando su rango con `bins` y etiquetas con `labels`.

 Donde, para contar cuántos elementos hay en cada subconjunto de datos definido, como lo es `Sexo` y `CategoriaIMC`; se utiliza la función `size()`, convirtiendolo en porcentaje al dividirlo entre el total de registros con `len(data)` y multiplicarlo por 100.

```python
data['CategoriaIMC']=pd.cut(
data['IMC'],
bins = [0, 18.5, 24.9, 29.9, 100],
labels = ['Bajo peso', 'Normalidad', 'Sobrepeso', 'Obesidad'],
)

print(data.groupby(['Sexo', 'CategoriaIMC'], observed=True).size()/len(data)*100)
```


De esta manera, es posible observar los porcentajes representativos de la variable `IMC`, en sus distintas categorías y subgrupos, de acuerdo a la variable `Sexo` .

Viéndose mayor incidencia de *Bajo peso* y *Normalidad* en mujeres, con un porcentaje de 8.10% y 7.20% respectivamente. Mientras que, las categorías *Sobrepeso y Obesidad* se ven dominadas por hombres, con un porcentaje de 15.68% y 23.35% respectivamente.





***Definición de una variable de interés para inferencia***

Por consiguiente, una variable de salida a considerar es `NivelDeObesidad`. La cual, siendo de naturaleza cualitativa clasificándose en sus respectivas categorías; funciona como indicador de la salud pública al estar directamente relacionado con el riesgo de enfermedades crónicas.

Además, al poder ser predicha o explicada mediante otras variables tales como`Edad`, `Sexo`, `Peso` o `IMC`, puede ser adecuada para inferencias en distintos casos. Sin embargo, su detalle se inclina a resultados categóricos, sin ofrecer detalle numérico.

De este modo, utilizando nuevamente  la función `value_counts()`será posible obtener un análisis descriptivo de la distribución de valores, utilizando `normalize=True` para visualizar las proporciones de presencia de cada sexo en los distintos niveles de obesidad.

```python
data['NivelDeObesidad'].value_counts(normalize=True)*100
data.groupby('Sexo')['NivelDeObesidad'].value_counts(normalize=True)*100
```

El análisis descriptivo demuestra que, de manera general, la mayor proporción de la población se encuentra en `Obesity_Type_III` en mujeres (30.97%)y `Obesity_Type_II` en hombres (27.62%).

Viéndose ligeramente bajo en `Overweight_Level_II` (9.88%) Y `Obesity_Type_II` (0.19%) para el caso de las mujeres. Mientras que, en el caso de los hombres varía entre `Insufficient_Weight`(9.27%) y `Obesity_Type_III` (0.09%).

  Permitiéndonos determinar que las categorías de común incidencia sin distinguir el sexo son: `Obesity_Type_I`, `Overweight_Level_I` y `Normal_Weight`.  

***Análisis gráfico enfocado en una subpoblación***

Para el análisis gráfico de la subpoblación *femenina* se evidenciará la relación entre la variable de salida `NivelDeObesidad` y la variable cuantitativa `Edad`.

Utilizando un gráfico de dispersión con la función `plt.scatter` mediante el modulo `pyplot` que cuenta con las funciones necesarias para crear gráficos.

```python
%matplotlib inline
import matplotlib.pyplot as plt

mujeres = data[data['Sexo'] == 'Female']

plt.scatter(x=mujeres['Edad'], y=mujeres['NivelDeObesidad'], alpha=0.5)
plt.xlabel("Edad")
plt.ylabel("Nivel de Obesidad")
plt.title("Relación entre Edad y Nivel de Obesidad en Mujeres")
plt.show()

```

En consecuencia, podemos observar que, dentro de la categoría `Normal_Weight`, `Insuffincient_Weight` y `Obesity_Type_III` su mayor concentración se da entre los 14 y 25 años aproximadamente. Mientras que categorías como `Overweight_Level_I`, `Overweight_Level_II` y`Obesity_Type_I` se distribuyen entre los 14 años hasta los 40 años aproximadamente; siendo la categoría menos común `Obesity_Type_II`.

El gráfico sugiere que las categorías relacionadas a obesidad se concentran entre los 20 y 40 años, con una disminución de casos a partir de los 45 años aproximadamente.

Indicando una amplia variabilidad en los rangos de edad y las distintas categorías, reflejando que la edad no necesariamente determina de manera uniforme el nivel de obesidad.

En conclusión, la obesidad depende de múltiples factores externos e internos, manifestándose de manera distinta en hombres y mujeres. Por ello, las asunciones respecto a tendencias por edad o género representan una alta variabilidad y limitada precisión.

Tomando en cuenta que, mediciones médicas, como el IMC son solo una perspectiva parcial del estado nutricional, ya que no hacen distinción de grasas corporal y masa muscular, lo que puede llevar a clasificaciones sesgadas de la población.

Donde, para mayor precisión serían pertinentes estudios más precisos. No obstante, mediciones similares a las realizadas pueden interpretarse como medidas preventivas útiles.
